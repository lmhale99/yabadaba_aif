# Discussion

This collects random thoughts and considerations associated with further progress 

## AIF vs AIF2

AIF uses numpy arrays for adsorp and desorp.  AIF2 uses pandas Dataframes for adsorp and desorp.

Going forward, one of these formats should be picked over the other. It is largely up to your personal preference.

- Time tests showed that loading AIF records is 2-3X faster than loading AIF2 records.
- The dataframe representations are more convenient for investigating the tabular data.
- Setting the values as separate arrays is slightly easier than building a dataframe, but not much.
- With a little extra work, AIF could be extended to generate the dataframe representation on request. The main concern is making it so changes to the dataframe values updates the array values, or simply stating that the dataframe values are copies.



## File size comparison

The JSON format that is generated uses a tree-like structure to group terms with the same prefix together, e.g. _exptl_adsorptive is at path exptl.adsorptive.  This reduces the overall file size for separate single-measurement records as each prefix only appears in a given record once.  Thus, JSON files can be of comparable or slightly smaller sizes than AIF files.


## Unofficial AIF terms

I defined the record values based on what I could find online and with example records.  If the recognized terms are standardized, then simply defining and updating the value definitions is enough.  However, if the schema is loose or allows for custom terms, they would need to be handled differently.

- Currently, unknown values are ignored during data transformations.
- Custom terms could be tracked with a little extra work by collecting them into a dict. They would lack most value features but would be retained during transformation.

## And the sprinkles on top

I did minimal effort in getting the above up and running, but could easily add or adjust capabilities if this seems like an interesting route to go down.

- The date fields you have are currently handled only as str but could be automatically interpreted by datetime objects.  This is very easy and something I've done before.
- yabadaba also has built-in unit conversion capabilities so loaded values could be automatically converted on the fly to common user-specified units.
- If you have an interest in having a CDCS curator with such data, I think it could be possible to create an XSLT that can automatically convert from the XML representation to the AIF format. It is a pain to do, but once it is done it is done.
- You can integrate Python-based plotting and analysis tools directly into the record objects.
